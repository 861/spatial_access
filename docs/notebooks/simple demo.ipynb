# Access Metrics (Origin Attributes)

```python
from spatial_access.Models import *
```

```python
# Access Score
```

```python
access = AccessModel(network_type='walk',
                     sp_matrix_filename=None,
                     sources_filename='data/input_data/tracts2010.csv',
                     destinations_filename='data/input_data/health_chicago.csv',
                     decay_function='logit')
```

```python
category_dict = {
"Hospitals": [10,10,10,10,10],
"Federally Qualified Health Centers": [8, 7, 6, 5, 4],
"School-Based Health Centers": [7, 7, 6, 6, 5],
"All Free Health Clinics": [5, 5, 5, 4, 4],
"Other Health Providers": [4,3,2,1,1]
}
```

```python
access.calculate(upper_threshold=1800, normalize=True,
                 category_weight_dict=category_dict,good_access_threshold=40)
```

```python
access.model_results.head()
```

```python
access.model_results.to_csv('access2010.csv')
```
